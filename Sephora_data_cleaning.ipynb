```python
import pandas as pd
```

```python
df_products = pd.read_csv("Sephora_product_review/product_info.csv")
df_products.head()
```

```python
df_products.columns
```

```python
columns_to_drop = ['child_count', 'child_max_price', 'child_min_price']
df_products_clean = df_products.drop(columns=columns_to_drop)
df_products_clean.head()
```

```python
df_products_clean.isnull().sum()
```

```python
# Fill 'size' and 'ingredients' with 'Unknown'
df_products_clean['size'] = df_products_clean['size'].fillna("Unknown")
df_products_clean['ingredients'] = df_products_clean['ingredients'].fillna("Unknown")

# Drop rows where price_usd is missing
df_products_clean = df_products_clean.dropna(subset=['price_usd'])

# Reset index
df_products_clean = df_products_clean.reset_index(drop=True)

df_products_clean.head()
```

```python
import os
import pandas as pd

# Folder where you stored the review files 
review_folder = "Sephora_product_review/reviews" 

# Get list of all CSV files in that folder
review_files = [f for f in os.listdir(review_folder) if f.endswith(".csv")]

# Load and merge all review CSVs
df_reviews = pd.concat([
    pd.read_csv(os.path.join(review_folder, file), dtype=str) for file in review_files
], ignore_index=True)

df_reviews.head()
```

```python
df_reviews = df_reviews.drop(columns=["Unnamed: 0"], errors="ignore")

df_reviews.head()
```

```python
df_reviews.isnull().sum()
```

```python
df_merged = df_reviews.merge(df_products_clean, on='product_id', how='inner')
df_merged.head()
```

```python
df_merged.shape
```

```python
df_products_clean.shape
```

```python
df_merged.columns.tolist()
```

```python
df_merged = df_merged.rename(columns={
    'rating_x': 'rating',               # user's review rating
    'rating_y': 'avg_rating'            # product's overall average rating
})

# Drop other unnecessary duplicated columns (if you havenâ€™t already)
df_merged = df_merged.drop(columns=[
    'product_name_x', 'brand_name_x', 'price_usd_x'
], errors='ignore')  # errors='ignore' just in case you've already dropped them

df_merged.head()
```

```python
df_merged.to_csv("cleaned_product_reviews.csv")
```
